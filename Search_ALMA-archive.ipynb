### This code will search ALMA archive for your objects based on coordinates
This ALMINER is somehow very slow and will take time ~1hrs for ALPINE 

```python
from astropy.table import Table
import numpy as np
import pandas as pd
from astropy.io import ascii
import warnings
import subprocess
import alminer
warnings.filterwarnings("ignore") 
```

### Read your catalog

```python
incat = './data_Duplication/ALPINE_merged_catalogs.fits'
tdata = Table.read(incat,format='fits')
id = tdata['name']
ra = tdata['RA']
dec = tdata['DEC']
```

### Search in ALMA archive

```python
### make Pandas DF
mycat = {
    "Name": id,
    "RAJ2000": ra,
    "DEJ2000": dec}

mycat = pd.DataFrame(mycat)
```

```python
# search radius (in arcmin) is set to 1arcsec
# Point = True for phase center search
myquery = alminer.catalog(mycat,search_radius=0.017,point=True,print_targets=True,public=None)
```

```python
# Save query output to CSV table
alminer.save_table(myquery, filename="ALPINE_plus")
```
