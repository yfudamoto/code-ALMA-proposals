I test here using <a href="https://cesam.lam.fr/a2c2s/data_release.php">ALPINE catalog</a>

```python
# This is using original checker

from astropy.table import Table
incat = './data_Duplication/ALPINE_merged_catalogs.fits'
tdata = Table.read(incat,format='fits')

# Assuming RA and Dec is in degree
objid = tdata['name']
ra = tdata['RA']
dec = tdata['DEC']

search_radius = 3 # arcsec
```

### Read ALMA Observations list
- Need "general_code/duplication-check-csv_mod.csv" in the current woring directory

```python
from astropy.table import Table
incat_ALMA = './data_Duplication/duplication-check-csv_mod.csv'
tdata_ALMA=  Table.read(incat_ALMA,format='csv')

Project_Code = tdata_ALMA['Project Code']
Target_Name = tdata_ALMA['Target Name']
RA_ALMA = tdata_ALMA['RA']
DEC_ALMA = tdata_ALMA['Dec']
Band = tdata_ALMA['Band']
Frequency = tdata_ALMA['Ref.Frequency']
Resolution = tdata_ALMA['Req. Ang. Res.']
```

```python
from astropy.coordinates import SkyCoord

coord_ALMA = SkyCoord(RA_ALMA,DEC_ALMA,frame='icrs',unit='deg')
coord_object = SkyCoord(ra,dec,frame='icrs',unit='deg')
```

### Here starts searching

```python
import numpy as np

search_radius = search_radius / 3600 #convert to deg

for idx in range(len(objid)):
    list= coord_ALMA.separation(coord_object[idx])
    arg = np.where(list.deg < search_radius)
    
    if len(arg[0]) == 0:
        print('There is no observations of :', objid[idx])
    else :
        for idx_match in range(len(arg[0])):
            # print(arg[0])
            # print(idx_match,arg[0][idx_match])
            print('There is match with Project ID: ',Project_Code[arg[0][idx_match]])
            print('\t Object name in the project:', Target_Name[arg[0][idx_match]], 'Coordinate: RA=', RA_ALMA[arg[0][idx_match]],'DEC=', DEC_ALMA[arg[0][idx_match]],'Band=',Band[arg[0][idx_match]], 'Ref. Frequency=', Frequency[arg[0][idx_match]], ' Ang. Resolution=',Resolution[arg[0][idx_match]])
```
